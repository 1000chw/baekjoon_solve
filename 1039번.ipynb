```python
import sys

N, K = map(int, sys.stdin.readline().split())
N = [int(i) for i in str(N)]
length_N = len(N)

if length_N == 1:
    print(-1)
    sys.exit()
if length_N == 2 and N[1] == 0:  # 두자리 (ex. 10)일 때만
    print(-1)
    sys.exit()

# 최댓값과 비교하자
max_N = sorted(N, reverse=True)  # N의 최댓값
result = []


# list 요소 바꾸기
def change(arr, x, y):
    tmp = arr[x]
    arr[x] = arr[y]
    arr[y] = tmp
    return arr


# list -> int
def ConvertToInt(arr):
    res = ''
    for i in arr:
        res += str(i)
    return int(res)


def solve(depth, arr):
    if depth == K:  # K번 바꾸면 종료
        result.append(ConvertToInt(arr))
        return

    index = -1  # 최댓값의 자릿수와 다른 인덱스
    for i in range(length_N):
        if arr[i] != max_N[i]:
            index = i
            break
    if index == -1:  # 최댓값이랑 똑같으면->
        if length_N != len(set(arr)):  # 겹치는 숫자가 있으면 원하는 값임
            print(ConvertToInt(arr))
            sys.exit()
        else:  # 숫자가 다 다르면 마지막 두 개만 바꾸면 됨
            arr = change(arr, length_N - 2, length_N - 1)
            solve(depth + 1, arr)
            return
    else:  # 최댓값이랑 다르다면 ->
        tmp_index = []
        # 최댓값으로 만들어줄 숫자 인덱스 찾기
        for i in range(index + 1, length_N):
            if max_N[index] == arr[i]:
                tmp_index.append(i)

        for i in tmp_index:
            arr1 = list(map(int, arr))
            solve(depth + 1, change(arr1, index, i))


solve(0, N)

# K번 거친 요소들 중 최댓값
max_result = 0
for i in result:
    if i > max_result:
        max_result = i
print(max_result)
```
